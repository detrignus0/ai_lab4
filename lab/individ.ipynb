```python
import matplotlib.pyplot as plt
import pandas as pd

csv_data = """Дата,Температура
2024-05-01,18
2024-05-02,20
2024-05-03,22
2024-05-04,21
2024-05-05,19"""

with open("temperature.csv", "w", encoding="utf-8") as f:
    f.write(csv_data)

df = pd.read_csv("temperature.csv")
df["Дата"] = pd.to_datetime(df["Дата"])
df.set_index("Дата", inplace=True)

plt.plot(df["Температура"])
plt.title("Изменение температуры")
plt.xlabel("Дата")
plt.ylabel("Температура")
plt.grid(True)
plt.show()
```
