```python
import pandas as pd

animals_list = ['cat', 'dog', 'rabbit', 'parrot', 'fish']
series = pd.Series(animals_list)

print("Исходный Series:")
print(series)

print("\nНачальный тип данных (.dtype):")
print(series.dtype)

categorical_series = series.astype('category')

print("\nSeries после преобразования:")
print(categorical_series)

print("\nНовый тип данных (.dtype):")
print(categorical_series.dtype)
```
