```python
import numpy as np
import pandas as pd

data = [1.2, np.nan, 3.4, np.nan, 5.6, 6.8]
series = pd.Series(data)

print("Исходный Series:")
print(series)

mean_value = series.mean()
filled_series = series.fillna(mean_value)

print("\nСреднее значение непустых элементов:")
print(mean_value)

print("\nSeries после заполнения пропусков:")
print(filled_series)
```
