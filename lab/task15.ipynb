```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

dates = pd.date_range(start="2024-03-01", periods=30, freq="D")
s = pd.Series(np.random.randint(50, 151, size=30), index=dates)

plt.plot(s)
plt.title("График значений")
plt.xlabel("Дата")
plt.ylabel("Значение")
plt.grid(True)
plt.show()
```
