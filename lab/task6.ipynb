```python
import numpy as np
import pandas as pd

data = [1.2, np.nan, 3.4, np.nan, 5.6, 6.8]
series = pd.Series(data)

print("Исходный Series:")
print(series)

nan_mask = series.isna()
nan_indices = series[nan_mask].index

print("\nЕсть ли пропущенные значения (NaN):")
print(nan_mask.any())

print("\nИндексы элементов с пропущенными значениями:")
print(nan_indices.tolist())
```
