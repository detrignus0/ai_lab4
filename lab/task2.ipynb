```python
import pandas as pd

indices = ['A', 'B', 'C', 'D', 'E']
values = [12, 24, 36, 48, 60]

series = pd.Series(data=values, index=indices)
print("Исходный Series:\n", series)

element_by_label = series.loc['C']
print("\nЭлемент с индексом 'C' (через .loc):", element_by_label)

element_by_position = series.iloc[2]
print("Третий элемент (через .iloc):", element_by_position)
```
